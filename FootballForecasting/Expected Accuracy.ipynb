```python
import numpy as np
import pandas as pd
import itertools
```

```python
def calculate_accuracy(row):
    home_accuracy=abs(row["actual_home"]-row["predicted_home"])/(1+row["predicted_home"])
    away_accuracy = abs(row["actual_away"]-row["predicted_away"])/(1+row["predicted_away"])
    accuracy=1-((home_accuracy+away_accuracy)/2)
    return accuracy
```

```python
home=[0,1,2,3]
away=[0,1,2,3]
predicted_home=[0,1,2,3]
predicted_away=[0,1,2,3]
possible_scores = pd.DataFrame(list(itertools.product(*[home,away,predicted_home,predicted_away])),columns=["actual_home","actual_away","predicted_home","predicted_away"])
```

```python
possible_scores
```

```python
possible_scores["accuracy"]=possible_scores.apply(calculate_accuracy,axis=1)
possible_scores
```

```python
data_home=possible_scores[possible_scores["actual_home"]>possible_scores["actual_away"]].copy()
data_away=possible_scores[possible_scores["actual_home"]<possible_scores["actual_away"]].copy()
data_draw=possible_scores[possible_scores["actual_home"]==possible_scores["actual_away"]].copy()
data_home_draw=possible_scores[possible_scores["actual_home"]>=possible_scores["actual_away"]].copy()
data_away_draw=possible_scores[possible_scores["actual_home"]<=possible_scores["actual_away"]].copy()
```
